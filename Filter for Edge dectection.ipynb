# Edge Dectection Filter


We can try out how to make an edge dectection filter by using open cv

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np

%matplotlib inline
```

```python
image = mpimg.imread('data/panda.jpeg')

plt.imshow(image)
```

First let us change this image to gray scale so that we can use a simple enough filter for the demonstration

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

Here we are going to create a kernel to perfrom the Sobel operation,
essentialy this allows us to dectect horizontal and vertical lines of high-frequency.

```python
sobel_y = np.array([[ -1, -2, -1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])
filtered_image = cv2.filter2D(gray, -1, sobel_y)

plt.imshow(filtered_image, cmap='gray')
```

```python
sobel_x = sobel_y.T
filtered_image = cv2.filter2D(gray,-1,sobel_x)
plt.imshow(filtered_image, cmap='gray')
```
